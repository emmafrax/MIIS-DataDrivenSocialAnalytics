```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_svmlight_file
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.utils import shuffle
import scipy
import time
from math import sqrt
```

WARNING: As the two programs for each Linear and Logistic Regression task use the same variables names, in order to execute them, the plots must be executed right after the main program. 

## LINEAR REGRESSION

### ABALONE

ABALONE DATASET: To start with this first task, we picked the Abalone regression dataset (#of features: 8 and#of data: 4,177). All the programming has been done with python, using the Jupyter notebookenvironment. The linear regression algorithm has been imported from the scikit-learn library. Forthe figures presented in this report, we applied a statistical approach, averaging over 250 iterations.Each of these iterations started by doing a random shuffle of the data.

```python
#We start loading our dataset defining both x_data and y_data.
x_data, y_data = load_svmlight_file("../datasets/abalone.txt")
#The algorithm that is gonna be used is defined. In this case, we are gonna use a linear model.
reg = linear_model.LinearRegression()
#The size of the x_data is determined to define some auxiliar variables.
s1,f = x_data.shape
#We generat a vector of dataset size N that changes depending on the total size of the used dataset.
h = int(s1/100)
N= np.arange(h,s1,h)

#We define how many times the simulation will be repeated, to obtain some statistics and smooth the final plots.
est=250

#We define some empty vectors to store the information we want to obtain.
error_f = np.zeros([N.size])
time_f = np.zeros([N.size])
error_all = np.zeros([N.size,est])
time_all = np.zeros([N.size,est])
weights_final = np.zeros([N.size,f])
weights = np.zeros([N.size,f])

#A first loop is defined to do the statistics.
for i in range(est):
    #Some other empty auxiliar variables are defined.
    error_t = []
    time_t=[]
    coef = []
    #We perform a shuffle of the data to use different intial data points.
    x_temp,y_temp = shuffle(x_data, y_data, random_state=i)
    k=0;
    #A second loop is started to perform the simulation for increasing size of training dataset.
    for j in N:
        start_time = time.perf_counter() #The initial start time is computed 
        x_train = x_temp[:j] #The training set is defined to be N-sized.
        y_train=y_temp[:j]
        reg.fit(x_train,y_train) #The linear regression is performed.
        y_pred = reg.predict(x_train)
        t_pr=time.perf_counter() - start_time #We compute the final computing 
        time_t= np.append(time_t,t_pr) #Both time and error vectors are updated.
        error_t = np.append(error_t, mean_squared_error(y_train,y_pred))
        weights[k,:] = reg.coef_ #We save the obtained coefficients at each iteration
        k=k+1; #We update the k variable to coun how many iterations have been performed.
    #For each iteration of the statistics we update both error and time matrix, so 
    #we can save up the data obtained for each simulation.    
    error_all[:,i] = error_t 
    time_all[:,i] = time_t 
    #We sum the weights, the error and the time data to compute the mean in the end.
    weights_final = weights_final + weights;
    error_f = error_f+error_t
    time_f = time_f+time_t
#We divide the mean vectors by the total number of repetitions that have been performed.
error_f = error_f/est; 
time_f = time_f/est;
weights_final = weights_final/est;
```

```python
#We define two empy vectors to save the variance values.
variance_error = np.zeros([N.size]);
variance_time = np.zeros([N.size]);
#A loop is performed to compute the variance of the result of each repetition once all simulations have already been performed.
for i in range(est):
    #For each simulation, the variance is computed and saved in a variance vector for both time and error.
    variance_time = variance_time + (time_all[:,i]-time_f)**2/est
    variance_error = variance_error + (error_all[:,i]-error_f)**2/est
```

```python
#The variance of CPU time is plotted.
plt.plot(N,variance_time)
plt.title('variance of CPU time (abalone dataset)')
plt.xlabel('Training set size')
plt.ylabel('variance')
plt.grid()
```

```python
#The variance of the error is plotted
plt.plot(N,variance_error)
plt.title('variance of MSE (abalone dataset)')
plt.xlabel('Training set size')
plt.ylabel('variance')
plt.grid()
```

```python
#The weights are plotted.
plt.plot(N,weights_final)
plt.title('Linear Regression: Weights behavior (abalone dataset)')
plt.xlabel('Training set size')
plt.ylabel('weights')
plt.grid()
```

```python
#The MSE of the linear regression is plotted.
plt.plot(N,error_f)
plt.title('Linear Regression: MSE (abalone dataset)')
plt.xlabel('Training set size')
plt.ylabel('MSE')
plt.grid()
```

```python
#The CPU time of the linear regression is plotted.
plt.plot(N,time_f)
plt.title('Linear Regression: CPU time (cpusmall dataset)')
plt.xlabel('Training set size')
plt.ylabel('CPU time')
plt.grid()
```
