### DIFFERENT SIZE OF TRAINING / TEST DATASETS

A first comparison of the different results obtained when training the MLP model with an increasig size of training set are performed in this first section of the notebook.

- TRAINING: 1/4
TEST: 3/4

```python
#W start loading our dataset defining both x_data and y_data.
x_data, y_data = load_svmlight_file("../datasets/mushrooms.txt")
s1,f = x_data.shape
size = np.int(s1/4)
x_train = x_data[:size]
y_train = y_data[:size]
x_test  = x_data[size:]
y_test  = y_data[size:]
```

```python
#A vector of alphas is dephined. To do so, an equally logspaced vector is generated from 10^-8 to 10^2.
alpha_vec = np.logspace(-8,2,100)

#Some void vectors are generated to update and save the error variable.
MSE_training_vec1 = [];
MSE_test_vec1 = [];

#A loop is generated to define for each alpha value the corresponding neural network model.
for a in alpha_vec:
    #A logistic Multi-Layer perceptron neural network model is defined.
    clf = MLPClassifier(activation = 'logistic', max_iter = 1000,solver='lbfgs',
                        alpha=a,hidden_layer_sizes=(60, 20), random_state=1,)
    #The data is fitted using the training dataset
    clf.fit(x_train,y_train)
    
    #A prediction is performed for both training and test datasets.
    y_pred_train = clf.predict(x_train)
    y_pred_test = clf.predict(x_test)

    #The corresponding error is performed
    MSE_train = mean_squared_error(y_train,y_pred_train)
    MSE_test = mean_squared_error(y_test,y_pred_test)

    #The values are stored for each alpha value.
    MSE_training_vec1.append(MSE_train);
    MSE_test_vec1.append(MSE_test);
```
